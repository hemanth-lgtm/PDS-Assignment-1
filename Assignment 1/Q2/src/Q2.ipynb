```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

# **Question 2**

**Five Visualizations for the Student Performance Dataset**

```python
#Load the dataset
student_data = pd.read_csv('StudentsPerformance.csv')
student_data.head()
```

### **Gender Distribution (Stacked Bar Plot)**

```python
# Stacked bar plot for gender distribution
gender_counts = student_data['gender'].value_counts()
gender_percentage = gender_counts / gender_counts.sum() * 100

plt.bar(gender_counts.index, gender_counts, color=['skyblue', 'salmon'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()

print("Gender Percentage:")
print(gender_percentage)
```

**The stacked bar plot provides a clear visual of the number of male and female students in the dataset. Itâ€™s particularly useful when comparing multiple categories within the same group, but in this case, it clearly shows the gender split. This visualization makes it easy to compare counts and visualize gender balance.**

### **Race/Ethnicity Distribution (Pie Chart)**

```python
# Pie chart for race/ethnicity distribution
race_counts = student_data['race/ethnicity'].value_counts()

plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Race/Ethnicity Distribution')
plt.show()

print("Race/Ethnicity Counts:")
print(race_counts)
```

**The pie chart offers a percentage breakdown of the student distribution across different race/ethnicity groups. It makes it easier to see the proportional representation of each group in the dataset, which can be useful for analyzing demographic diversity.**

### **Parental Education vs Average Scores (Point Plot)**

```python
# Point plot for parental education vs average score
sns.pointplot(x='parental level of education', y='average_score', data=student_data, palette='Set2')
plt.title('Parental Level of Education vs Average Scores')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Test Scores')
plt.xticks(rotation=45)
plt.show()
```

**The point plot shows the average scores for students based on parental education level, with points representing the mean for each group. This visualization is useful for identifying trends or differences in performance based on parental education, and itâ€™s easier to compare exact values than in a box or strip plot.**

### **Reading Score vs. Math Score (Scatter Plot with Regression Line)**


```python
# Scatter plot with regression line for Reading Score vs Math Score
sns.regplot(x='reading score', y='math score', data=student_data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Reading Score vs Math Score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.show()
```

**This visualization helps assess whether thereâ€™s a correlation between reading and math scores, allowing for exploration of relationships between subjects.**

**Race/Ethnicity vs Average Scores (Clustered Bar Plot)**

```python
# Calculate average scores for each race/ethnicity
average_scores = student_data.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().reset_index()

# Plotting the clustered bar plot
average_scores.plot(x='race/ethnicity', kind='bar', stacked=False, color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Scores')
plt.xticks(rotation=45)
plt.legend(title='Subjects')
plt.show()
```

**This visualization provides insights into how average scores differ among racial/ethnic groups, allowing for analysis of equity and representation in academic performance.**
